```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset:
italian_football = pd.ExcelFile('Italian Football.xlsx')
sa = pd.read_excel(italian_football, 'Serie A TIM')

# Displaying the last 5 champions:
sa.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
sa.shape
```

```python
# Displaying the datatypes presented in the columns:
sa.info()
```

```python
# Displaying statistical data of the numerical data present in the dataset:
sa.describe()
```

```python
# Displaying number of titles by nation:
sa['Competition name'].value_counts().plot(kind = 'pie', figsize = (8, 8))
plt.ylabel(None)
plt.title('Competition name', fontweight = 'bold')
plt.show()
```

```python
# Displaying winners:
sa['Winners'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 6))
plt.ylabel(None)
plt.title('Winners', fontweight = 'bold')
plt.show()
```

```python
# Displaying runners-up:
sa['Runners-up'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 10))
plt.ylabel(None)
plt.title('Runners-up', fontweight = 'bold')
plt.show()
```

```python
# Displaying winners:
sa['Third Place'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 8))
plt.ylabel(None)
plt.title('Third Place', fontweight = 'bold')
plt.show()
```

```python
# Displaying top scorers:
sa['Top scorer 1'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 20))
plt.ylabel(None)
plt.title('Top scorers', fontweight = 'bold')
plt.show()
```

```python
# Displaying clubs with most top scorers:
sa['Club 1'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 6))
plt.ylabel(None)
plt.title('Clubs with most top scorers', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.figure(figsize = (8, 4))
plt.title('Top scorer amount of goals by season', fontweight = 'bold')
plt.plot(sa['Season'], sa['Number of goals 1'], color = 'cyan')
plt.show()
```
