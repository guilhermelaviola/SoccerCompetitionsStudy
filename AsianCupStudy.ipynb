```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import seaborn as sns
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT_tournaments_finals.xlsx')
ac = pd.read_excel(nt_tournaments, 'Asian Cup finals')

# Displaying the last 5 champions:
ac.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ac.shape
```

```python
# Displaying the datatypes presented in the columns:
ac.info()
```

```python
# Displaying number of titles by nation:
ac.groupby('Year').first()['Winners'].value_counts().plot(kind = 'barh', color = 'gold')
plt.ylabel(None)
plt.title('Winners by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by nation:
ac.groupby('Year').first()['Runners-up'].value_counts().plot(kind = 'barh', color = 'silver')
plt.ylabel(None)
plt.title('Runners-up by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying host countries by nation:
ac.groupby('Year').first()['Host Country'].value_counts().plot(kind = 'barh', color = 'lime')
plt.ylabel(None)
plt.title('Host Countries by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying host cities that hosted finals:
ac.groupby('Year').first()['Host City'].value_counts().plot(kind = 'barh', color = 'lime')
plt.ylabel(None)
plt.title('Host Cities', fontweight = 'bold')
plt.show()
```

```python
# Displaying venues that hosted finals:
ac.groupby('Year').first()['Venue'].value_counts().plot(kind = 'barh', color = 'magenta')
plt.ylabel(None)
plt.title('Venues', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.title('Attendances by Year', fontweight = 'bold')
plt.plot(ac['Year'], ac['Attendance'], color = 'orange')
plt.show()
```

```python
# Displaying the distribution of the way the match was won:
ac.groupby('Year').first()['How the match was won'].value_counts().plot(kind = 'pie')
plt.ylabel(None)
plt.title('How the matches were won', fontweight = 'bold')
plt.show()
```
