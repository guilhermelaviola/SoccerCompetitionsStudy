```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import seaborn as sns
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT tournaments finals.xlsx')
ca = pd.read_excel(nt_tournaments, 'Copa America finals')

# Displaying the last 5 champions:
ca.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ca.shape
```

```python
# Displaying the datatypes presented in the columns:
ca.info()
```

```python
# Displaying winners by Nation:
ca.groupby('Year').first()['Winners'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Winners by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying runners-up by Nation:
ca.groupby('Year').first()['Runners-up'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Runners-up by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying the host countries by nation:
ca.groupby('Year').first()['Host country'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Host countries by nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying Venues used for the 1st games and how many times they were used:
ca.groupby('Year').first()['Venue (1st game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Venues used (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying Venues used for the 2nd games and how many times they were used:
ca.groupby('Year').first()['Venue (2nd game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Venues used (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying Venues used for the 3rd games and how many times they were used:
ca.groupby('Year').first()['Venue (3rd game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Venues used (3rd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying cities that hosted the tournment in the 1st game and how many times they did it:
ca.groupby('Year').first()['Host city (1st game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Host city (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying cities that hosted the tournment in the 2nd game and how many times they did it:
ca.groupby('Year').first()['Host city (2nd game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Host city (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying cities that hosted the tournment in the 3rd game and how many times they did it:
ca.groupby('Year').first()['Host city (3rd game)'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Host city (3rd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the distribution of how the matches were won:
ca.groupby('Year').first()['How the match was won'].value_counts().plot(kind = 'pie')
plt.ylabel(None)
plt.title('How the matches were won', fontweight = 'bold')
plt.show()
```

```python
# Displaying the distribution of the tournament name as it changes thoughout the years:
ca.groupby('Year').first()['Tournament name'].value_counts().plot(kind = 'pie')
plt.ylabel(None)
plt.title('Tournament name', fontweight = 'bold')
plt.show()
```
