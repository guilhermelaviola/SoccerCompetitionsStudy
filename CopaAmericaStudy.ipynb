```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT_tournaments_finals.xlsx')
ca = pd.read_excel(nt_tournaments, 'Copa America finals')

# Displaying the last 5 champions:
ca.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ca.shape
```

```python
# Displaying the datatypes presented in the columns:
ca.info()
```

```python
# Displaying the number of titles by Country:
ca['Winners'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Winners', fontweight = 'bold')
plt.show()
```

```python
# Displaying the number of Runners-up by Country:
ca['Runners-up'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Runners-up', fontweight = 'bold')
plt.show()
```

```python
# Displaying the host countries by Country:
ca['Host country'].value_counts().plot(kind = 'barh', color = 'orange', figsize = (8, 8))
plt.ylabel(None)
plt.title('Host Countries', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues used for the finals (1st leg):
ca['Venue (1st game)'].value_counts().plot(kind = 'barh', color = 'orange', figsize = (8, 8))
plt.ylabel(None)
plt.title('Venues (1st leg)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues used for the finals (2nd leg):
ca['Venue (2nd game)'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Venues (2nd leg)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues used for the finals (3rd leg):
ca['Venue (3rd game)'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Venues (3rd leg)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities that hosted the finals (1st leg):
ca['Host city (1st game)'].value_counts().plot(kind = 'barh', color = 'orange', figsize = (8, 6))
plt.ylabel(None)
plt.title('Host city (1st leg)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities that hosted the finals (2nd leg):
ca['Host city (2nd game)'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Host city (2nd leg)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities that hosted the finals (3rd leg):
ca['Host city (3rd game)'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Host city (3rd leg)', fontweight = 'bold')
plt.show()
```

```python
# Checking if the club won Copa America at home or away:
def champion_at_home(row):
  winners_country = row['Winners']
  host_country = row['Host country']
  if winners_country == host_country:
    result = 'Home'
  else:
    result = 'Away'
  return result

# Applying the function above on a column of the dataset and generating a new column as result:
ca['Home champion?'] = ca.apply(champion_at_home, axis=1)
ca.head()
```

```python
# Displaying the clubs who were champions at home and away:
ca['Home champion?'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Champions at home and away', fontweight = 'bold')
plt.show()
```

```python
# Displaying the distribution of the way the matches were won:
ca['How the match was won'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('How the matches were won', fontweight = 'bold')
plt.show()
```

```python
# Displaying the distribution of the tournament name as it changed thoughout the years:
ca['Tournament name'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Tournament name', fontweight = 'bold')
plt.show()
```
