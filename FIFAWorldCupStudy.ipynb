```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT_tournaments_finals.xlsx')
wc = pd.read_excel(nt_tournaments, 'FIFA World Cup finals')

# Displaying the last 5 champions:
wc.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
wc.shape
```

```python
# Displaying the datatypes presented in the columns:
wc.info()
```

```python
# Displaying the number of titles by country:
wc['Winners'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Winners by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying the number of runners-up by country:
wc['Runners-up'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Runners-up by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.figure(figsize = (12, 8))
plt.title('Attendances by Year', fontweight = 'bold')
plt.plot(wc['Year'], wc['Attendance'], color = 'green')
plt.show()
```

```python
# Displaying how the games were won:
wc['How the game was won'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('How the games were won', fontweight = 'bold')
plt.show()
```
