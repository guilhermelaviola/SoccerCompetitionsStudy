```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT_tournaments_finals.xlsx')
wc = pd.read_excel(nt_tournaments, 'FIFA World Cup finals')

# Displaying the last 5 champions:
wc.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
wc.shape
```

```python
# Displaying the datatypes presented in the columns:
wc.info()
```

```python
# Displaying general statistical information of the number data present in the dataset:
wc.describe()
```

```python
# Displaying the number of titles by country:
wc['Winners'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 4))
plt.ylabel(None)
plt.title('Winners', fontweight = 'bold')
plt.show()
```

```python
# Displaying the number of runners-up by country:
wc['Runners-up'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 6))
plt.ylabel(None)
plt.title('Runners-up', fontweight = 'bold')
plt.show()
```

```python
# Displaying distribution of venues that hosted finals:
wc['Venue'].value_counts().plot(kind = 'barh',color = 'cyan', figsize = (8, 8))
plt.ylabel(None)
plt.title('Venues that hosted finals', fontweight = 'bold')
plt.show()
```

```python
# Displaying distribution of cities that hosted finals:
wc['Host city'].value_counts().plot(kind = 'barh',color = 'cyan', figsize = (8, 8))
plt.ylabel(None)
plt.title('Cities that hosted finals', fontweight = 'bold')
plt.show()
```

```python
# Displaying distribution of countries that hosted finals:
wc['Host country'].value_counts().plot(kind = 'barh',color = 'cyan', figsize = (8, 8))
plt.ylabel(None)
plt.title('Countries that hosted finals', fontweight = 'bold')
plt.show()
```

```python
# Checking if the club won UEL at home or away:
def champion_at_home(row):
  winners_country = row['Winners']
  host_country = row['Host country']
  if winners_country == host_country:
    result = 'Home'
  else:
    result = 'Away'
  return result

# Applying the function above on a column of the dataset and generating a new column as result:
wc['Home champion?'] = wc.apply(champion_at_home, axis=1)
wc.head()
```

```python
# Displaying the clubs who were champions at home and away:
wc['Home champion?'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Champions at home and away', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.figure(figsize = (16, 8))
plt.title('Attendances by Year', fontweight = 'bold')
plt.plot(wc['Year'], wc['Attendance'], color = 'cyan')
plt.show()
```

```python
# Displaying how the games were won:
wc['How the game was won'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('How the games were won', fontweight = 'bold')
plt.show()
```
