```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import seaborn as sns
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT tournaments finals.xlsx')
wc = pd.read_excel(nt_tournaments, 'FIFA World Cup finals')

# Displaying the last 5 champions:
wc.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
wc.shape
```

```python
# Displaying the datatypes presented in the columns:
wc.info()
```

```python
# Displaying number of titles by nation:
wc.groupby('Year').first()['Winners'].value_counts().plot(kind = 'barh',
                                                                     figsize = (8, 8))
plt.ylabel(None)
plt.title('Winners by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runenrs-up by nation:
wc.groupby('Year').first()['Runners-up'].value_counts().plot(kind = 'barh',
                                                                     figsize = (8, 8))
plt.ylabel(None)
plt.title('Runners-up by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.title('Attendances by Year', fontweight = 'bold')
plt.plot(wc['Year'], wc['Attendance'])
plt.show()
```

```python
# Displaying the distribution of the way the match was won:
wc.groupby('Year').first()['How the game was won'].value_counts().plot(kind = 'pie',
                                                                     figsize = (8, 8))
plt.ylabel(None)
plt.title('How the match was won', fontweight = 'bold')
plt.show()
```
