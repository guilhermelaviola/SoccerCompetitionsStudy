```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing the dataset:
nt_tournaments = pd.ExcelFile('NT_tournaments_finals.xlsx')
ac = pd.read_excel(nt_tournaments, 'Africa Cup of Nations finals')

# Displaying the last 5 champions:
ac.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ac.shape
```

```python
# Displaying the datatypes presented in the columns:
ac.info()
```

```python
# Displaying general statistical information of the number data present in the dataset:
ac.describe()
```

```python
# Displaying the number of titles by nation:
ac['Winners'].value_counts().plot(kind = 'barh', color = 'purple', figsize = (8, 8))
plt.ylabel(None)
plt.title('Winners', fontweight = 'bold')
plt.show()
```

```python
# Displaying the number of runners-up by nation:
ac['Runners-up'].value_counts().plot(kind = 'barh', color = 'purple', figsize = (8, 8))
plt.ylabel(None)
plt.title('Runners-up', fontweight = 'bold')
plt.show()
```

```python
# Displaying the host countries by nation:
ac['Host country'].value_counts().plot(kind = 'barh', color = 'purple', figsize = (8, 8))
plt.ylabel(None)
plt.title('Host Countries', fontweight = 'bold')
plt.show()
```

```python
# Displaying the host cities that hosted finals:
ac['Host city'].value_counts().plot(kind = 'barh', color = 'purple', figsize = (8, 8))
plt.ylabel(None)
plt.title('Host Cities', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues that hosted finals:
ac['Venue'].value_counts().plot(kind = 'barh', color = 'purple', figsize = (8, 12))
plt.ylabel(None)
plt.title('Venues', fontweight = 'bold')
plt.show()
```

```python
# Checking if the team won African Cup of Nations at home or away:
def champion_at_home(row):
  winners_country = row['Winners']
  host_country = row['Host country']
  if winners_country == host_country:
    result = 'Home'
  else:
    result = 'Away'
  return result

# Applying the function above on a column of the dataset and generating a new column as result:
ac['Home champion?'] = ac.apply(champion_at_home, axis=1)
ac.head()
```

```python
# Displaying the clubs who were champions at home and away:
ac['Home champion?'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Champions at home and away', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances by year:
plt.figure(figsize = (8, 4))
plt.title('Attendances', fontweight = 'bold')
plt.plot(ac['Year'], ac['Attendance'], color = 'purple')
plt.show()
```

```python
# Displaying the distribution of the way the match was won:
ac['How the match was won'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('How the matches were won', fontweight = 'bold')
plt.show()
```
