```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from google.colab import drive
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
drive.mount('/content/drive')
```

```python
# Setting the search parameters
filename = 'italian-football.xlsx'
search_path = '/content/drive/My Drive/Datasets/Soccer'
# Search for the file
file = None
for root, dirs, files in os.walk(search_path):
    if filename in files:
        file= os.path.join(root, filename)
        break
    df = pd.read_excel(file)
```

```python
# Importing the dataset:
italian_football = pd.ExcelFile(file)
ci = pd.read_excel(italian_football, 'Coppa Italia')

# Displaying the last 5 champions:
ci.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ci.shape
```

```python
# Displaying the datatypes presented in the columns:
ci.info()
```

```python
# Displaying number of titles by club:
ci['Winners'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 4))
plt.ylabel(None)
plt.title('Winners by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by club:
ci['Runners-up'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 6))
plt.ylabel(None)
plt.title('Runners-up by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of times the venues were used in the 1st games:
ci['Venue (1st game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 4))
plt.ylabel(None)
plt.title('Venues (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of times the venues were used in the 2nd games:
ci['Venue (2nd game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 3))
plt.ylabel(None)
plt.title('Venues (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities where the 1st games took place in:
ci['City (1st game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 3))
plt.ylabel(None)
plt.title('Cities (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities where the 2nd games took place in:
ci['City (2nd game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 3))
plt.ylabel(None)
plt.title('Cities (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances of the 1st and 2nd leg matches by season:
plt.figure(figsize = (8, 4))
plt.plot(ci['Year'], ci['Attendance (1st game)'])
plt.plot(ci['Year'], ci['Attendance (2nd game)'])
```
