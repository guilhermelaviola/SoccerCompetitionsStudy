```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import seaborn as sns
```

```python
# Importing the dataset:
italian_football = pd.ExcelFile('Italian Football.xlsx')
ci = pd.read_excel(italian_football, 'Coppa Italia')

# Displaying the last 5 champions:
ci.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ci.shape
```

```python
# Displaying the datatypes presented in the columns:
ci.info()
```

```python
# Displaying number of titles by club:
ci.groupby('Season').first()['Winners'].value_counts().plot(kind = 'barh',
                                                            color = 'gold')
plt.ylabel(None)
plt.title('Winners by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by club:
ci.groupby('Season').first()['Runners-up'].value_counts().plot(kind = 'barh',
                                                               color = 'silver')
plt.ylabel(None)
plt.title('Runners-up by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of times the venues were used in the 1st games:
ci.groupby('Season').first()['Venue (1st game)'].value_counts().plot(kind = 'barh',
                                                                     color = 'orange')
plt.ylabel(None)
plt.title('Venues (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of times the venues were used in the 2nd games:
ci.groupby('Season').first()['Venue (2nd game)'].value_counts().plot(kind = 'barh',
                                                                     color = 'lime')
plt.ylabel(None)
plt.title('Venues (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities where the 1st games took place in:
ci.groupby('Season').first()['City (1st game)'].value_counts().plot(kind = 'barh',
                                                                    color = 'magenta')
plt.ylabel(None)
plt.title('Cities (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the cities where the 2nd games took place in:
ci.groupby('Season').first()['City (2nd game)'].value_counts().plot(kind = 'barh',
                                                                    color = 'blue')
plt.ylabel(None)
plt.title('Cities (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances of the 1st and 2nd leg matches by season:
plt.figure(figsize = (16,8))
plt.plot(ci['Season'], ci['Attendance (1st game)'])
plt.plot(ci['Season'], ci['Attendance (2nd game)'])
```
