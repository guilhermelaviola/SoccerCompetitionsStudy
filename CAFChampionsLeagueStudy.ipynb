```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import seaborn as sns
```

```python
# Importing the dataset:
caf = pd.ExcelFile('CAF Competitions finals.xlsx')
ccl = pd.read_excel(caf, 'CAF Champions League finals')

# Displaying the last 5 champions:
ccl.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ccl.shape
```

```python
# Displaying the datatypes presented in the columns:
ccl.info()
```

```python
# Displaying number of titles by club:
ccl.groupby('Season').first()['Winners'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Breakdown of winners by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by club:
ccl.groupby('Season').first()['Runners-up'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Breakdown of runners-up by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of titles by country:
ccl.groupby('Season').first()['Winners country'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Breakdown of winners by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by country:
ccl.groupby('Season').first()['Runners-up country'].value_counts().plot(kind = 'barh')
plt.ylabel(None)
plt.title('Breakdown of runners-up by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying the attendances of the 1st and 2nd leg matches by season:
plt.figure(figsize = (16,8))
plt.plot(ccl['Season'], ccl['Attendance (1st game)'])
plt.plot(ccl['Season'], ccl['Attendance (2nd game)'])
```

```python
# Displaying the total attendances (1st and 2nd leg matches) by season:
plt.figure(figsize = (16,8))
plt.plot(ccl['Season'], ccl['Attendance (1st game)'] + ccl['Attendance (2nd game)'])
```

```python
# Displaying the average attendances (1st and 2nd leg matches) by season:
plt.figure(figsize = (32,16))
plt.plot(ccl['Season'], (ccl['Attendance (1st game)'] + ccl['Attendance (2nd game)']) / 2)
```
