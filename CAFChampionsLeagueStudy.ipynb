```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from google.colab import drive
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
drive.mount('/content/drive')
```

```python
# Setting the search parameters
filename = 'caf-competitions-finals.xlsx'
search_path = '/content/drive/My Drive/Datasets/Soccer'
# Search for the file
file = None
for root, dirs, files in os.walk(search_path):
    if filename in files:
        file= os.path.join(root, filename)
        break
    df = pd.read_excel(file)
```

```python
# Importing the dataset:
caf = pd.ExcelFile(file)
ccl = pd.read_excel(caf, 'CAF Champions League finals')

# Displaying the last 5 champions:
ccl.tail()
```

```python
# Displaying the number of rows and columns in the dataset:
ccl.shape
```

```python
# Displaying the datatypes presented in the columns:
ccl.info()
```

```python
# Displaying number of titles by club:
ccl['Winners'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 8))
plt.ylabel(None)
plt.title('Winners by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by club:
ccl['Runners-up'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 8))
plt.ylabel(None)
plt.title('Runners-up by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of titles by country:
ccl['Winners country'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 4))
plt.ylabel(None)
plt.title('Winners by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying number of runners-up by country:
ccl['Runners-up country'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 6))
plt.ylabel(None)
plt.title('Runners-up by Country', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues (1st game):
ccl['Venue (1st game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 10))
plt.ylabel(None)
plt.title('Venues (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the venues (2nd game):
ccl['Venue (2nd game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 10))
plt.ylabel(None)
plt.title('Venues (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying host cities (1st game):
ccl['Host city (1st game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 6))
plt.ylabel(None)
plt.title('Host cities (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying host cities (2nd game):
ccl['Host city (2nd game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 8))
plt.ylabel(None)
plt.title('Host cities (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying host countries (1st game):
ccl['Host country (1st game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 4))
plt.ylabel(None)
plt.title('Host countries (1st game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying host countries (2nd game):
ccl['Host country (2nd game)'].value_counts().plot(kind = 'barh', color = 'green', figsize = (8, 4))
plt.ylabel(None)
plt.title('Host countries (2nd game)', fontweight = 'bold')
plt.show()
```

```python
# Displaying the distribution how the match was won:
ccl['How the match was won'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('How the match was won', fontweight = 'bold')
plt.show()
```
