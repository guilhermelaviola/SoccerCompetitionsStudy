```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from google.colab import drive
sns.set(style = 'whitegrid')
plt.style.use('seaborn-pastel')
drive.mount('/content/drive')
```

```python
# Setting the search parameters
filename = 'italian-football.xlsx'
search_path = '/content/drive/My Drive/Datasets/Soccer'
# Search for the file
file = None
for root, dirs, files in os.walk(search_path):
    if filename in files:
        file= os.path.join(root, filename)
        break
    df = pd.read_excel(file)
```

```python
# Importing the datasets:
italian_football = pd.ExcelFile(file)
cp = pd.read_excel(italian_football, 'Campionato Primavera')
cip = pd.read_excel(italian_football, 'Coppa Italia Primavera')
```

# **Campionato Primavera**

```python
# Displaying champions:
cp['Champion'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 6))
plt.ylabel(None)
plt.title('Champions', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions cities:
cp['City'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 6))
plt.ylabel(None)
plt.title('Cities', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions regions:
cp['Region'].value_counts().plot(kind = 'barh', color = 'cyan', figsize = (8, 4))
plt.ylabel(None)
plt.title('Regions', fontweight = 'bold')
plt.show()
```
